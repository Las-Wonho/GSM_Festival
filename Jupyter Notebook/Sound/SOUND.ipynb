```python
class node():
    def init(self):
        self.H=1
        self.dataSet=np.arange(0,10,0.1)
        self.ZERO=1e-15
        self.INF=1e15
    def calc(self,data):
        return data*self.H
    def cost(self,label):
        count=0
        for i in range(0,len(label)):
            count+=(label[i]-calc(self.dataSet[i]))
        return count
    def vis(self,label):
        plt.plot(self.dataSet,label)
        plt.show()
```

```python
n=node()
n.init()
```

```python
n.vis(np.arange(0,100,1))
```
