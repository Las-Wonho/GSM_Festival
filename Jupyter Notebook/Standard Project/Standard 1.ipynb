```python
def cost(anser,hyp):
    return np.square(anser-hyp)
```

```python
cost(1,3)
```

```python
def learningRate():
    return 0.001
def costFuntion(h):
    a=0
    ba=0
    for i in np.arange(-100,100,0.1):
        a+=cost(GoodNode(i),Node(h,i,b))
    for i in np.arange(-100,100,0.1):
        ba+=cost(GoodNode(i),Node(learningRate()+h,i,b))
    if a==0:
        return 0
    else :
        return ba-a
```

```python
costFuntion(3)
```

```python
costNumber=-1
H=1
while costNumber != 0:
    costNumber=costFuntion(H)
    print(costNumber)
    print("s"+str(H))
    H-=(int(costNumber)*0.000001)
    if(int(costNumber)==0 and costNumber!=0):
        if(costNumber<0):
            H+=0.000001
        else:
            H-=0.000001
```

<h1>It's langueges's limite. and real number in cimputer is not perfect</h1>
<br><h1>a new function</h1>

```python
def NewCostFuntion(h):
    a1=0
    a2=0
    b=2
    for i in np.arange(-100,100,0.1):
        a1+=cost(GoodNode(i),Node(h+learningRate(),i,b))
    for i in np.arange(-100,100,0.1):
        a2+=cost(GoodNode(i),Node(h-learningRate(),i,b))
    result=(a1-a2)/2
    if abs(result)<ZERO:
        return 0
    else:
        return result
```

```python
NewCostFuntion(3.00001)
```

```python
costN=1
H=600
while costN!=0:
    costN=NewCostFuntion(H)
    if (costN/100000)<ZERO*100:
        H=H-(costN/10000)
    else:
        H=H-(costN/1000000)
    print(costN)
```
